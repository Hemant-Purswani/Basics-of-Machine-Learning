```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#importing numpy, pandas, seaborn- for data visualisation, matplotlib- for plotting 
```

```python
#plotting a scatter plot
x= np.arange(0,10)
y= np.arange(11,21)
plt.scatter(x,y,c="r",cmap="magenta")
plt.xlabel("x-axis" , fontsize=14)
plt.ylabel("y-axis", fontsize=14)

plt.savefig("figure1.png")
```

```python
#scatter plot for various markers and subplot
x= np.arange(0,10)
y= np.arange(11,21)
plt.subplot(2,2,1)
plt.scatter(x,y,c="r",marker="D")
plt.xlabel("x-axis" , fontsize=11)
plt.ylabel("y-axis", fontsize=11)
plt.subplot(2,2,2)
plt.scatter(x,y,c="b",marker="s")
plt.xlabel("x-axis" , fontsize=11)
plt.ylabel("y-axis", fontsize=11)
plt.subplot(2,2,3)
plt.scatter(x,y,c="g",marker="P")
plt.xlabel("x-axis" , fontsize=11)
plt.ylabel("y-axis", fontsize=11)
plt.subplot(2,2,4)
plt.scatter(x,y,c="black",marker="*")
plt.xlabel("x-axis" , fontsize=11)
plt.ylabel("y-axis", fontsize=11)
```

```python
#Plotting a barchart
x= [2,8,10]
y=[11,16,9]
plt.bar(x,y,color="magenta")
plt.xlabel("x-axis" , fontsize=11)
plt.ylabel("y-axis", fontsize=11)
```

```python
#Boxplot 
#There is an outlier
#Right Skewed 
x =[3,4,5,7,9,8,12,13,7,8,19,90,12,15]
plt.boxplot(x,vert=False)
```

```python
#Reading the csv file
a=pd.read_csv("seeds.csv")
a.head()

```

```python
#describe provides us with the calculation of the statistical data (Label wise)
a.describe()
```

```python
#from the the 'Type' column it is counting the number - type wise 
a["Type"].value_counts()
```

```python
plt.bar([1,2,3],a["Type"].value_counts())
```

```python
#plotting a barchart from the data given by the value_counts() function 
a["Type"].value_counts().plot.bar()
```

```python
#Scattering of the Kernel.Width and Kernel.Length is done to find the type of association
plt.scatter(a["Kernel.Width"],a["Kernel.Length"])
plt.xlabel("Kernel.Width")
plt.ylabel("Kernel.Length")
print()
print(""" Positive Association, as Kernel-Width increases Kernel_Length also increases 
""")
```

```python
#countplot takes the attribute which we want to plot in string, data=equal to the name of your data frame
#seaborn is better than matplotlib in terms of visualisation
sns.countplot('Type',data=a,color=(0.6,0.4,0.2))
```

```python
sns.scatterplot(a["Kernel.Width"],a["Kernel.Length"],data=a, hue="Type", palette="crest")
# Different Color is assigned to different types of seed
#The Kernel Length and Kernel Width of type 2 is the highest 
```

```python
sns.jointplot(x="Perimeter", y="Compactness", data=a)
#seed having perimeter 14 have a higher compactness
```

```python
plt.boxplot(a["Perimeter"])
```

```python
sns.boxplot("Type", "Compactness", data=a, palette="rainbow")
#Type 2 has higher median
#Type 1 has outliers
```

```python
sns.pairplot(data=a,hue="Type")
#Asymmetric
```

```python
sns.violinplot("Type","Compactness", data=a )
#Similar as boxplot
```

```python
sns.kdeplot("Type","Compactness", data=a )
#looks like top view of 3D model
```

```python
pd.plotting.andrews_curves(a,class_column="Type" )
#Type 1 and 2 difficult to differentiate
```
