```python
cars_data=pd.read_csv("Toyota.csv",na_values=["??","????"])
cars_data.head(5)
#while reading it replaces the values of ?? and ???? with nan
```

```python
cars_data=cars_data.replace("three",3)
cars_data=cars_data.replace("four",4)
cars_data=cars_data.replace("five",5)
#replacing letter values with numbers
```

```python
cars_data.head()
```

```python
#Q.13
cars_data["Doors"]=cars_data.Doors.astype('int64')

cars_data["MetColor"]=cars_data["MetColor"].astype('object')

cars_data["Automatic"]=cars_data["Automatic"].astype('object')

cars_data.info()

#changing data types of columns
```

```python
#df.dropna helps in removing missing values 
cars_data=cars_data.dropna()
cars_data.head()
```

```python
adults['workclass']=adults['workclass'].astype('category')
adults['education']=adults['education'].astype('category')
  
  
#Assigning numerical values and storing it in another columns
adults['workclass_new']=adults['workclass'].cat.codes
adults['education_new']=adults['education'].cat.codes 
```

```python
adults.head()
```

```python
def label_encoding(column_name, data_frame):
  a=np.unique(data_frame[column_name])
  b=list(np.arange(0,len(a)))
  print(a,b)
  for i in b:
    data_frame[column_name]=data_frame[column_name].replace(a[i],b[i])
  
  return data_frame


a=input("Enter the name of the feature in which you want to do label-encoding: ")
df4=label_encoding(a,adults)
df4.head()
```
